```python
# pandas_ta.atr?
def atr(inp):
    atr_series = pandas_ta.atr(high=inp['High'], low=inp['Low'], close=inp['Close'], length=14)
    return atr_series.sub(atr_series.mean()).div(atr_series.std())
```

4. calculatiing  montly returns for diffrent time horizons

```python
def cal_return(df): 
    outliers_cutoff = 0.005
    lag_periods = [1, 2, 3, 6, 9, 12]
    for lag in lag_periods:
        df[f'return_{lag}m'] = (df['Adj Close'].pct_change(lag)
                                .pipe(lambda x: x.clip(lower=x.quantile(outliers_cutoff),
                                                      upper=x.quantile(1 - outliers_cutoff)))
                                .add(1).pow(1/lag).sub(1))
    return df
```

```python
top150=top150.groupby(level=1,group_keys=False).apply(cal_return).dropna()
```
