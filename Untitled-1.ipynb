reading the data from wikipedia to know about sp500 list

```python
sp500=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

storing the symbol and handling errors in symbol

```python
symbol_list=sp500['Symbol'].replace('.','-').tolist()
symbol_list
```

```python
end_date='2023-09-27'
start_date=pd.to_datetime(end_date)-pd.DateOffset(365*8)
start_date
```

downloading data to get started

```python
df=yf.download(tickers=symbol_list, start=start_date, end=end_date )
```
